```python
from PIL import Image, ImageOps
import pytesseract
import numpy as np

filename = '01.png'
img1 = Image.open(filename).convert('L')
img1 = ImageOps.invert(img1)
img1 = img1.point(lambda x: 0 if x < 128 else 255, '1')
img1 = np.array(img1)

text = pytesseract.image_to_string(img1, lang='eng', config='--psm 6')
print(text)
```
